```python
import pandas as pd 
import numpy as np
from sklearn.datasets import load_breast_cancer

# Aqui eu carreguei os dados brutos do DataSet

dados_cancer = load_breast_cancer()
```

```python
# Aqui eu criei a tabela com as colunas certas

df = pd.DataFrame(dados_cancer.data, columns=dados_cancer.feature_names)

# Adicionando a coluna de resposta (target)

df['diagnostico'] = dados_cancer.target
```

```python
print("Informações dos dados:")
df.info()
```

```python
nulos = df.isnull().sum().sum()
print(f"Total de dados nulos: {nulos}")
```

```python
duplicadas = df.duplicated().sum()
print(f"Total de linhas duplicadas: {duplicadas}\n")
```

```python
porcentagem = df['diagnostico'].value_counts(normalize=True) * 100 # Aqui nós verificamos a porcentagem de pacientes com casos benignos e malignos de cancêr
print("\nPorcentagem na base de dados:")
print(porcentagem)
```

```python
print("Distribuição dos diagnósticos:") 
print(df['diagnostico'].value_counts()) # O comando "value_counts()" irá nós possibilitar quantos pacientes possuem cancêr benigno (o que está com indice 1)
print("-" * 30)                         # E quantos possuem cancêr maligno (o que está com indice 0)
```

```python
print("\nEstatísticas descritivas resumidas (Média, Mín, Máx):")
print(df.iloc[:, :5].describe()) # O comando "decribe()" calcula sozinho a média, desvio padrão, valor mínimo e máximo de todas as colunas
```

```python
df = df.drop(['id', 'nome_paciente'], axis=1) # Este comando permite a remoção de alguma coluna, no caso estamos excluindo a coluna de Id que é algo inútil
                                              # Porém, está coluna não existe em nosso DataSet, logo não irá servir de nada e irá ocasionar em um erro
```

```python
df.to_csv('cancer_de_mama_limpo.csv', index=False) # Criação de um novo CSV para o "novo" DataSet limpo
```

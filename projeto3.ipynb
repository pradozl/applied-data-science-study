```python
import pandas as pd

df = pd.read_csv('Planilha todas as cidade ESAVI.xlsx - Página1.csv')

print("Tamanho do DataSet (Linhas, Colunas):", df.shape)
```

```python
df['Classificação de gravidade'] = df['Classificação de gravidade'].str.replace(r'^\d+:\s*', '', regex=True)
```

```python
df['Data da Notificação'] = pd.to_datetime(df['Data da Notificação'], format='mixed', dayfirst=False)

# Converte a data, para ficar mais padronizado
```

```python
colunas_remover = ['Gravidade', 'Mês de gestação', 'Mulher amamentando no momento da vacinação?', 'Gestante']
df_limpo = df.drop(columns=colunas_remover)

# Resolvi retirar a coluna 'Gravidade', pois ela se torna redundante, visto que existe a coluna 'Classificação de gravidade'
# Resolvi retirar a coluna 'Mês de gestação' pois quase todas as respostas estão marcadas como 0 ou está vazio
# Resolvi retirar a coluna 'Mulher amamentando no momento da vacinação?' pois apenas existem 3 respostas como 'Sim', o que torna meio irrelevante
# Resolvi retirar a coluna 'Gestante' pois praticamente todas as respsotas estão marcadas como 'Não' ou 0
```

```python
colunas_para_limpar = [
    'Imunobiológico (vacina)', 'Dose', 'Tipo de Evento', 
    'Reação / evento adverso', 'Classificação de gravidade', 
    'Desfecho (evolução do caso)', 'Tipo de Atendimento'
]

# Remove "Número:" e espaços extras no começo e fim
for col in colunas_para_limpar:
    if col in df_limpo.columns:
        # regex=True permite que o Python entenda o padrão "digito seguido de dois pontos"
        df_limpo[col] = df_limpo[col].astype(str).str.replace(r'\d+:\s*', '', regex=True).str.strip()
```

```python
df_limpo['Idade Evento'] = df_limpo['Idade Evento'].astype('Int64')

# Converter a idade para inteiro
```

```python
df_limpo.to_csv('dataset_ESAVI_limpo.csv', index=False, encoding='utf-8-sig')

print("As primeiras 5 linhas do seu novo dataset:")
display(df_limpo.head())

print(f"\nO dataset agora tem {df_limpo.shape[0]} linhas e {df_limpo.shape[1]} colunas.")
```

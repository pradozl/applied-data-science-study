```python
import pandas as pd 
from sklearn.datasets import load_breast_cancer

dados_cancer = load_breast_cancer()
df = pd.DataFrame(dados_cancer.data, columns=dados_cancer.feature_names)
df['diagnostico'] = dados_cancer.target

df.to_csv("breast_cancer_data.csv", index = False)

print("Primeiras 5 linhas do DataSet: ")
display(df.head())
print("Descrição e Informações das Colunas: ")
df.info()
```
